```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_set/lasvegas_tripadvisor.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Score'].value_counts()
```

### Box plot

```python
sns.catplot(
    x = 'Pool',
    y = 'Score',
    data = df,
    kind = 'box'
)
plt.show()
```

```python
sns.set(font_scale = 1.4)
sns.set_style('whitegrid')

sns.catplot(
    x = 'Pool',
    y = 'Score',
    data = df,
    kind = 'box'
)

plt.show()
```

```python
sns.set(font_scale = 1.25)
sns.set_style('darkgrid')

sns.catplot(
    x = 'Traveler type',
    y = 'Helpful votes',
    data = df,
    kind = 'box'
)

plt.show()
```

### Bar plot

```python
df['Traveler type'].value_counts().plot.bar();
```

```python
sns.set(font_scale = 1.25)
sns.set_style('whitegrid')


sns.catplot(
    x = 'Traveler type',
    y = 'Score',
    data = df,
    kind = 'bar',
    ci = False
)

plt.show()
```

```python
df['Traveler type'] = df['Traveler type'].astype('category')
df['Traveler type'].cat.categories
```

```python
sns.set(font_scale = 1.25)
sns.set_style('whitegrid')


sns.catplot(
    x = 'Traveler type',
    y = 'Score',
    data = df,
    kind = 'bar',
    ci = False
)

plt.show()
```

```python
sns.set(font_scale = 1.25)
sns.set_style('whitegrid')


sns.catplot(
    x = 'Traveler type',
    y = 'Score',
    data = df,
    hue = 'Tennis court',
    palette = sns.color_palette('tab10'),
    kind = 'bar'
)

plt.show()
```

### Point and count plots

```python
sns.catplot(
    x = 'Spa',
    y = 'Score',
    data = df,
    kind = 'point'
);
```

```python
sns.catplot(
    x = 'Spa',
    y = 'Score',
    data = df,
    kind = 'point',
    hue = 'Tennis court',
    dodge = True
);
```

### Additional catplot() options

```python
sns.set(font_scale = 1.4)
sns.set_style('whitegrid')
ax = sns.catplot(
        x = 'Traveler type',
        y = 'Score',
        col = 'User continent',
        col_wrap = 3,
        kind = 'bar',
        palette = sns.color_palette('Set2'),
        data = df
        )
ax.fig.suptitle('Hotel Score by Traveler Type and User Continent')
plt.subplots_adjust(top = .9)
ax.set_axis_labels('Traveler Type', 'Score')
plt.show()
```
